```python
import os
```

```python
pasta_pdf = r'caminho_pasta_pdf'
arquivo_txt = r'caminho_txt'
#with fecha os arquivos automaficamente, open abre os arquivos, r mofo leitura(read) as f é como o arquivo passa a ser chamado 
with open(arquivo_txt, "r", encoding="utf-8") as f:
    novos_nomes = [linha.strip() for linha in f.readlines()] #linhas.strip remove espaços extras, e le linha por linha e transofrma ela em uma nova variavel de lista
    pdfs = [arquivo for arquivo in os.listdir(pasta_pdf) if arquivo.lower().endswith(".pdf")] #os.listdir(pasta_pdf) divide os arquivos da pasta, .endswith(".pdf") verirfica se é pdf
    pdfs.sort()
    if len(pdfs) != len(novos_nomes): #Verificando se a quantidade bate
        print('A quantidade de PDFs e nomes TXT é diferente')
    else: 
        for pdf, novo_nome in zip(pdfs, novos_nomes):
            caminho_antigo = os.path.join(pasta_pdf, pdf) #junta os diretorios 
            caminho_novo = os.path.join(pasta_pdf, novo_nome + ".pdf")
            os.rename(caminho_antigo, caminho_novo) #renomeando 
    print('Arquivos renomeados com sucesso!')
```
